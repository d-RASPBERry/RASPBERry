```python
import os
import ray
import zlib
import json
import pickle
import random
import numpy as np
from pympler import asizeof
import gymnasium as gym
from dynaconf import Dynaconf
from tqdm.notebook import tqdm
from utils import check_path, convert_np_arrays, flatten_dict, env_creator
from ray.tune.logger import JsonLogger
from algorithms_with_statistics.ddqn_pber import DDQNWithMPBERAndLogging
from replay_buffer.mpber_ram_saver import MultiAgentPrioritizedBlockReplayBuffer
from ray.tune.registry import register_env
```

```python
env = gym.make("Pong")
asizeof_list = []
obs, _ = env.reset()
for x in range(10):
    obs = []
    for i in range(32):
        _ = env.step(env.action_space.sample())
        if _[2] or _[3]:
            _ = env.reset()
        obs.append(_[0])
    obs = np.array(obs)
    new_obs = zlib.compress(obs, 5)
    asizeof_list.append([asizeof.asizeof(obs), asizeof.asizeof(new_obs), asizeof.asizeof(new_obs)/ asizeof.asizeof(obs)])
asizeof_list = np.array(asizeof_list)
print(1 - asizeof_list.mean(0)[-1])
```
